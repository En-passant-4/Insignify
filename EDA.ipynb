```python
# Z_CostContact and Z_Revenue are constant values dropping them
df.drop(["Z_CostContact", "Z_Revenue"], inplace=True, axis=1)
df.head(10)
```

```python
# Encoding the education levels using an oridnal encoder

categories = [['Basic', 'Graduation', '2n Cycle', 'Master', 'PhD']] #Order matters.
encoder = OrdinalEncoder(categories=categories)

df['Education'] = encoder.fit_transform(df[['Education']])
```

```python
df.Marital_Status.unique()
```

```python
# Handle missing values 
df.fillna(df.median(numeric_only=True), inplace=True)

df['DaysSinceEnrollment'] = (pd.to_datetime('today') - df['Dt_Customer']).dt.days

# Count unique values in categorical columns
categorical_cols = ['Education', 'Marital_Status']
for col in categorical_cols:
    print(f"\nValue Counts for {col}:")
    print(df[col].value_counts())

# Distribution of income
plt.figure(figsize=(8, 5))
sns.histplot(df['Income'].dropna(), bins=30, kde=True)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Count')
plt.show()


df_new = df.drop(['Marital_Status'], axis=1)
# Correlation heatmap to identify relationships
plt.figure(figsize=(12, 6))
sns.heatmap(df_new.corr(), cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()

# Check response rate for campaigns
campaign_cols = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']
campaign_response = df[campaign_cols].mean()

plt.figure(figsize=(8, 5))
sns.barplot(x=campaign_response.index, y=campaign_response.values, palette='viridis')
plt.title('Campaign Response Rates')
plt.ylabel('Proportion of Customers Accepted')
plt.xticks(rotation=45)
plt.show()

# Analyze website activity
plt.figure(figsize=(8, 5))
sns.histplot(df['NumWebVisitsMonth'], bins=15, kde=True)
plt.title('Website Visits Distribution')
plt.xlabel('Number of Web Visits Per Month')
plt.ylabel('Count')
plt.show()

# Customer enrollment over time
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df['YearEnrolled'] = df['Dt_Customer'].dt.year
plt.figure(figsize=(8, 5))
sns.countplot(x=df['YearEnrolled'], palette='coolwarm')
plt.title('Customer Enrollment Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Customers')
plt.show()

df.drop(["Dt_Customer"], inplace=True, axis=1)
```

```python
# Create a new column for household composition category
df["Household_Comp"] = df["Kidhome"].astype(str) + " kids, " + df["Teenhome"].astype(str) + " teens"

# Count occurrences of each household composition type
household_counts = df["Household_Comp"].value_counts()

# Creating the treemap
plt.figure(figsize=(10, 6))
squarify.plot(sizes=household_counts.values, label=household_counts.index, 
              alpha=0.7, color=plt.cm.Paired.colors)

# Improve visualization
plt.title("Household Composition Treemap", fontsize=14)
plt.axis("off")  # Hide axes for better display

# Show plot
plt.show()
```

```python
# Aggregate total purchases per channel
channel_totals = {
    "Deals": df["NumDealsPurchases"].sum(),
    "Catalog": df["NumCatalogPurchases"].sum(),
    "Store": df["NumStorePurchases"].sum(),
    "Web": df["NumWebPurchases"].sum()
}

# Convert to DataFrame
channel_df = pd.DataFrame(list(channel_totals.items()), columns=["Channel", "Total Purchases"])

# Create a colorful stacked bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x="Channel", y="Total Purchases", data=channel_df, palette="coolwarm")

# Improve visualization
plt.title("Customer Purchases by Channel", fontsize=14)
plt.xlabel("Purchase Channel")
plt.ylabel("Total Purchases")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot
plt.show()
```

```python
# Define product spending columns
product_cols = ["MntWines", "MntMeatProducts", "MntFruits", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]

# Group by Generation and calculate average spending
generation_spending = df.groupby("Generation")[product_cols].mean().reset_index()

# Melt the dataframe for visualization
df_melted = generation_spending.melt(id_vars=["Generation"], var_name="Product", value_name="Avg_Spending")

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x="Generation", y="Avg_Spending", hue="Product", palette="coolwarm")
plt.title("Average Spending on Different Products by Generation")
plt.ylabel("Average Spending")
plt.xlabel("Generation")
plt.legend(title="Product Category")
plt.xticks(rotation=45)
plt.show()
#2. Heatmap: Spending Preferences by Generation
plt.figure(figsize=(10, 5))
sns.heatmap(generation_spending.set_index("Generation"), cmap="coolwarm", annot=True, fmt=".0f")
plt.title("Spending Pattern by Generation")
plt.ylabel("Generation")
plt.xlabel("Product Category")
plt.show()
```

```python
# Recency Analysis
recency_insights = {
    "Average Recency (Days Since Last Purchase)": df["Recency"].mean(),
    "Median Recency": df["Recency"].median(),
    "Most Common Recency Value (Mode)": df["Recency"].mode()[0],
    "Customers with High Recency (> 90 days) (%)": (df[df["Recency"] > 90].shape[0] / df.shape[0]) * 100,
    "Customers with Low Recency (< 30 days) (%)": (df[df["Recency"] < 30].shape[0] / df.shape[0]) * 100,
    "Optimal Recency for Campaign Response": df.groupby("Recency")["Response"].mean().idxmax(),
}

# Display results
for key, value in recency_insights.items():
    print(f"{key}: {value:.2f}" if isinstance(value, float) else f"{key}: {value}")
```
